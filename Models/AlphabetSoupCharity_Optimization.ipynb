## Step 1: Data Preprocessing
- Import libraries
- Load the dataset
- Drop unnecessary columns
- Handle rare categorical values
- Encode categorical variables
- Split into features and target arrays
- Split into training and testing datasets
- Apply feature scaling

```python
import pandas as pd
import tensorflow as tf
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import os
```

```python
application_df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
```

```python
application_df = application_df.drop(columns=["EIN", "NAME"])

application_count = application_df['APPLICATION_TYPE'].value_counts()

application_types_to_replace = list(application_count[application_count<100].index)
for app in application_types_to_replace:
    application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(app, "Other")

classification_counts = application_df["CLASSIFICATION"].value_counts()

classifications_to_replace = list(classification_counts[classification_counts<100].index)
for cls in classifications_to_replace:
    application_df["CLASSIFICATION"] = application_df["CLASSIFICATION"].replace(cls, "Other")

application_df = pd.get_dummies(application_df, dtype=float)

X = application_df.drop("IS_SUCCESSFUL", axis=1).values
y= application_df["IS_SUCCESSFUL"].values

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)

scaler = StandardScaler()
X_scaler = scaler.fit(X_train)
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Step 2: Model 1 - Initial Optimization
- Build Model 1
- Compile Model 1
- Train Model 1

```python
model_1 = tf.keras.models.Sequential()
model_1.add(tf.keras.layers.Dense(units=128, activation="relu", input_dim=X_train_scaled.shape[1]))
model_1.add(tf.keras.layers.Dense(units=64, activation="relu"))
model_1.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

model_1.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])

history_1 = model_1.fit(X_train_scaled, y_train, epochs=100)
```

## Step 3: Model 2 - Additional Layer
- Build Model 2
- Compile Model 2
- Train Model 2

```python
model_2 = tf.keras.models.Sequential()
model_2.add(tf.keras.layers.Dense(units=128, activation="relu", input_dim=X_train_scaled.shape[1]))
model_2.add(tf.keras.layers.Dense(units=64, activation="relu"))
model_2.add(tf.keras.layers.Dense(units=32, activation="relu"))
model_2.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

model_2.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])

history_2 = model_2.fit(X_train_scaled, y_train, epochs=150)
```

## Step 4: Model 3 - Different Activation Functions
- Build Model 3
- Compile Model 3
- Train Model 3

```python
from tensorflow.keras.layers import Dropout

model_3 = tf.keras.models.Sequential()

model_3.add(tf.keras.layers.Dense(units=64, activation="tanh", input_dim=X_train_scaled.shape[1]))
model_3.add(Dropout(0.3))

model_3.add(tf.keras.layers.Dense(units=32, activation="relu"))
model_3.add(Dropout(0.3))

model_3.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

model_3.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])

history_3 = model_3.fit(X_train_scaled, y_train, epochs=100)
```

## Step 5: Results Comparison
- Evaluate all three models
- Compare accuracies

```python
loss_1, acc_1 = model_1.evaluate(X_test_scaled, y_test)
loss_2, acc_2 = model_2.evaluate(X_test_scaled, y_test)
loss_3, acc_3 = model_3.evaluate(X_test_scaled, y_test)

print(f"Model 1 Accuracy: {acc_1}")
print(f"Model 2 Accuracy: {acc_2}")
print(f"Model 3 Accuracy: {acc_3}")
```
